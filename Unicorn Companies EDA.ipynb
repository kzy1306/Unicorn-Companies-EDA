## Results and evaluation


The `Date Joined` column does not just have the year—it has the year, month, and date, By extracting the year component from the `Date Joined` column and adding those year components into a new column to keep track of each company's year joined.

```python
companies["Year Joined"] = companies["Date Joined"].dt.year
companies.head()
```

```python
companies['Valuation'].describe()
companies['Industry'].value_counts()
companies[['Country/Region']].value_counts().head(10).plot(kind = 'bar', figsize = (12,5))
```

**Bar Plot for Unicorn Company founded throughout the years**

```python
plt.figure(figsize=(12, 6))
companies['Date Joined'].dt.year.value_counts().sort_index().plot(kind='bar')
plt.title('Number of Unicorn Companies Founded Over Time')
plt.xlabel('Year Joined')
plt.ylabel('Number of Companies')
plt.show()
```

**Unicorn Companies by Industries**

```python
plt.figure(figsize=(11, 6))  
sns.countplot(y='Industry', data=companies, order=companies['Industry'].value_counts().index)
plt.title('Unicorn Companies by Industry')
plt.show()
```

### Visualizing the time taken for unicorn status


**Using sample of the data**

```python
companies_sample = companies.sample(n = 50, random_state = 30)
```

```python
#Creating column for years taken to reach unicorn
companies_sample["years_till_unicorn"] = companies_sample["Year Joined"] - companies_sample["Year Founded"]

#Grouping Industry and years_till_unicorn for visualisation 
grouped = companies_sample[["Industry", "years_till_unicorn"]].groupby("Industry").max().sort_values(by="years_till_unicorn")
grouped
```

```python
plt.figure(figsize= (12, 5))
plt.bar(grouped.index, grouped["years_till_unicorn"])
plt.title("Bar plot of maximum years taken by company to become unicorn per industry")
plt.xlabel("Industry")
plt.ylabel("Maximum number of years")
plt.xticks(rotation=45, horizontalalignment='right')
plt.show()
```

**Question:** What do you observe from this bar plot?

- This bar plot shows that for this sample of unicorn companies, the largest value for maximum time taken to become a unicorn occurred in the Heath and Fintech industries, while the smallest value occurred in the Consumer & Retail industry.

### Visualizing the maximum unicorn company valuation per industry

```python
#creating a new column for visualizations

companies_sample['valuation_billions'] = companies_sample['Valuation']

companies_sample['valuation_billions'] = companies_sample['valuation_billions'].str.replace('$', '')

companies_sample['valuation_billions'] = companies_sample['valuation_billions'].str.replace('B', '')
# Converting column to type int
companies_sample['valuation_billions'] = companies_sample['valuation_billions'].astype('int')
companies_sample.head()
```

Now, prepare the data for modeling by performing the necessary calculations on it.

```python
grouped = (companies_sample[["Industry", "valuation_billions"]]
           .groupby("Industry")
           .max()
           .sort_values(by="valuation_billions")
          )
grouped
```

```python
plt.figure(figsize = (12,6))
plt.bar(grouped.index, grouped["valuation_billions"])
plt.title("Bar plot of maximum unicorn company valuation per industry (from sample)")
plt.xlabel("Industry")
plt.ylabel("Maximum valuation in billions of dollars")
plt.xticks(rotation=45, horizontalalignment='right');
```

**Question:** What do you observe from this bar plot? 

- This bar plot shows that for this sample of unicorn companies, the highest maximum valuation occurred in the Artificial Intelligence industry, while the lowest maximum valuation occurred in the Auto & transportation, and Consumer & retail industries.
